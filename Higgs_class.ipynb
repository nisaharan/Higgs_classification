```python
from sklearn.model_selection import train_test_split as tts
x = data_1.drop([0],axis=1).values   # independant features
y = data_1[0].values					       # dependant variable
x1 = pd.DataFrame(x)
y1 = pd.DataFrame(y)
x_train, x_test, y_train, y_test = tts(x1, y1, train_size = .8)
```

```python
# Define the model.
model = keras.models.Sequential([
    keras.layers.Dense(8, input_dim=28, activation='linear', kernel_initializer='uniform'),
    keras.layers.Dense(4, input_dim=8, activation='linear', kernel_initializer='uniform'),
    keras.layers.Dense(1, activation='tanh', kernel_initializer='normal')
])
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['accuracy'])
#model.compile(
    #optimizer='adam',
    #loss='mean_squared_error',
    #metrics=['accuracy'])
# Here different activations are tried with different no of layers. 
```

```python
model.summary()
```

```python
model.fit(
    x_train,
    y_train,
    epochs=100, batch_size=20, validation_split = .17, validation_batch_size = 25)
```

```python
# make predictions for the input data
#yhat = model.predict(x_test)
predictions = model.predict(x_test, batch_size=16, verbose=0)
```
